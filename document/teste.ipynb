## SIMULAÇÃO DO JOGO DE PERSEGUIÇÃO

        game_simulation(self, graph: Dgraphs)
Na nossa simulação, a policia e o ladão se movem em um grafo, onde os vértices representam locais e as arestas representam caminhos entre esses locais. 

O jogo se inicia com os policiais fazendo as patrulhas tradicionais(andando de 1 em 1 vertice), apois o a primeira patrulha dos policiais, o ladrão faz seu primeiro movimento entrando no local do roubo.A partir disso, a persiguiçao começa, onde a policia entre em modo perseguiçao(andando de 2 em 2 vertices), a policia possui a posiçao do ladão e assim procurnado o menor caminho ate ele.

O ladrão tenta escapar para um dos vértices de saída, enquanto a policia irá tentar tenta interceptá-lo. 

A cada etapa, o ladrão pode escolher um caminho para se mover, vendo os melhores caminhos ate os portos.

O jogo continua até que o ladrão seja capturado, diretamente pelos policiais ou fique encurralado, ou consiga escapar. 

Durante a simulação, resgistrado o histórico de movimentos do ladrão e da polícia, com os 'thief_log' e 'police_log', bem como o número de etapas necessárias para alcançar ou a vitória ou derrota.
